```python
import json, os
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

sns.set(style="whitegrid")
```

```python
stats = {}
with open("stats.json", "r") as f:
    stats = json.loads(f.read())


show = {'stats': ['blue_seat_per', 'blue_vote_per', 'red_vote_per', 'red_seat_per', 'red_scandal', 'blue_scandal']}
for title, group in show.items():
    if len(group) < 1:
        continue
    min_p = min(len(stats[l]) for l in group)
    min_p = min(1000, min_p)
    plot_data = {}
    for label in group:
        #plot_data[label] = stats[label][-min_p:]
        plot_data[label] = stats[label][:1000]
    data = pd.DataFrame(plot_data)
    fig = plt.figure(figsize=(16,9))
    ax = sns.lineplot(data=data, hue="event", style="event", dashes=False, markers=False, linewidth=1.5)
    ax.xaxis.set_major_locator(plt.MaxNLocator(5))
    ax.set_title(title)
    plt.show()
```
